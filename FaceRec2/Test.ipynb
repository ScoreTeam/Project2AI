```python
import h5py
import os
from pathlib import Path
```

```python

home=os.getenv("DEEPFACE_HOME", default=str(Path.home()))

file_name = "arcface_weights.h5"
output = home + "/.deepface/weights/" + file_name

# f = h5py.File(output, 'r')
# # dset = f['mydataset']
# print(f)

def printname(name):
    print(name)

with h5py.File(output, 'r') as f:
    f.visit(printname)
```

```python
def print_structure(name, obj):
    print(name)

def extract_weights(file_path, layer_name):
    with h5py.File(file_path, 'r') as f:
        f.visititems(print_structure)  # Print the structure of the file for debugging
        for name in f:
            if layer_name in name:
                layer_group = f[name]
                print(f"Layer group found: {name}")
                print(f"Contents of the layer group '{name}':")
                for k, v in layer_group.items():
                    print(f"Key: {k}, Type: {type(v)}")
                    if isinstance(v, h5py.Group):
                        print(f"Nested group: {k}")
                        for sub_key, sub_value in v.items():
                            print(f"    Sub Key: {sub_key}, Sub Type: {type(sub_value)}")
                            weights = {f"{k}/{sub_key}": sub_value[()] for sub_key, sub_value in v.items()}
                    else:
                        weights = {k: v[()] for k, v in layer_group.items()}  # Handle datasets directly in the layer group
                return weights
    return None

```

```python
conv1_weights = extract_weights(output,'conv1_conv')
print(conv1_weights)
```
